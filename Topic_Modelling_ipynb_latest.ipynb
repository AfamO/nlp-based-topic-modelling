# scrape using python and beautiful soup

```python
# Define a web-scraper function:

import requests
from bs4 import BeautifulSoup

def webScrapper(url,html_tag):
  reqObject = requests.get(url).text
  beautifier = BeautifulSoup(reqObject);
  title = beautifier.title.contents[0];
  print("Title::", title);
  #print("Text::", beautifier.text);
  #print(beautifier.prettify())
  tag_contents = beautifier.findAll(html_tag);

  textLists = [sentence.text for sentence in tag_contents];
  finalContents = " ".join(textLists).strip().replace("\xa0","").replace(">","").replace("<","").replace("\n","");
  return [title,finalContents];
```

## Test the function and scrape some articles

### import pandas dataframe

```python
import pandas as pd
```

```python
# Initialize some articles
corpus = [];
corpusDict = {};
dictArr = [];
counter = 0;
for url in articleLinks:
  result = webScrapper(url,"p");
  texts = result[1];
  corpusDict = {};
  corpus.append(texts);
  corpusDict["url"] = url;
  corpusDict["title"] = result[0];
  corpusDict["text"] = texts;
  dictArr.append(corpusDict);
  print("Counter ==",counter," current url==",url);
  counter = counter + 1;
corpusPD = pd.DataFrame(dictArr);
corpusPD.to_csv("fintechScrapedCorpus.csv");
corpus
```

# Import Spacy Lib for Data Pre-processing

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
processed_corpus = []
for curr_doc in corpus:
    doc = nlp(curr_doc)
    print("spacy doc =>", doc);
    token_lemma = [token_lemma.lemma_ for token_lemma in doc if not token_lemma.is_stop and not token_lemma.is_punct
                   and not token_lemma.is_space and not token_lemma.is_currency and token_lemma.is_alpha
                   and token_lemma.pos_ in ['NOUN', 'VERB', 'ADJ', 'ADV']]
    print("token lemma =>", token_lemma)
    article = " ".join(token_lemma)
    processed_corpus.append(article);
print("final processed corpus =>", processed_corpus)
```

# create document matrix

```python
dictionary = gensim.corpora.Dictionary([doc.split() for doc in processed_corpus])
doc_term_matrix = [dictionary.doc2bow(doc.split()) for doc in processed_corpus]
doc_term_matrix
```

## Now we can fit our model


### Let's use Latent Semantic Analysis (LSA) First

```python
from gensim.models import LsiModel as LSI
#LSA Model

lsa = LSI(doc_term_matrix, id2word=dictionary, num_topics=3)
```

```python
print(lsa.print_topics(num_topics=3, num_words=3))
```

## Let's try with LDA

```python
from gensim.models import LdaModel as LDA
#LDA Model

lda = LDA(doc_term_matrix, id2word=dictionary, num_topics=3, passes=10)
```

```python
# print result
print(lda.print_topics(num_topics=3, num_words=4))
```

```python
print("Topics Identified:")
for idx, topic in lda.print_topics(num_words=3):
    print(f"Topic {idx + 1}: {topic}")
```

# Calculate Coherence Score

```python
from gensim.models import CoherenceModel

# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=lda, texts=processed_corpus,
                                     dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('Coherence Score:', coherence_lda)
```
